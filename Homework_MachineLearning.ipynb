```python
import pandas as pd
url='insurance.csv'
data= pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
```

```python
from sklearn import linear_model 
model= linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['charges']),y=data['age'])
print(model.intercept_)
print(model.coef_)
```

```python
model.score(X=pd.DataFrame(data['charges']),y=data['age'])
```

```python
data.plot(kind='scatter',x='charges',y='age',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['charges']))
plt.plot(data['charges'],y_pred,color='b')
```

```python
model= linear_model.LinearRegression()
X=pd.DataFrame([data['charges'],data['charges']**2,data['charges']**3,data['charges']**4]).T
model.fit(X,y=data['age'])
print('Intercept của mô hình:',model.intercept_)
print('Coefient của mô hình:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X,y=data['age']))
```

```python
import numpy as np
data.plot(kind='scatter',x='charges',y='age',figsize=(9,9),color='black')
range=np.arange(10000,50000,1)
predictor=pd.DataFrame([range,range**2,range**3,range**4]).T
y_pred= model.predict(X=predictor)
plt.plot(range,y_pred,color='b')
```

```python
url='abalone.csv'
data= pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
```

```python
from sklearn import linear_model 
model= linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Whole weight']),y=data['Shucked weight'])
print(model.intercept_)
print(model.coef_)
```

```python
model.score(X=pd.DataFrame(data['Whole weight']),y=data['Shucked weight'])
```

```python
data.plot(kind='scatter',x='Whole weight',y='Shucked weight',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['Whole weight']))
plt.plot(data['Whole weight'],y_pred,color='b')
```

```python
model= linear_model.LinearRegression()
X=pd.DataFrame([data['Whole weight'],data['Whole weight']**2,data['Whole weight']**3,data['Whole weight']**4,data['Whole weight']**5,data['Whole weight']**6,data['Whole weight']**7,data['Whole weight']**8,data['Whole weight']**9,data['Whole weight']**10]).T
model.fit(X,y=data['Shucked weight'])
print('Intercept của mô hình:',model.intercept_)
print('Coefient của mô hình:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X,y=data['Shucked weight']))
```

```python
import numpy as np
data.plot(kind='scatter',x='Whole weight',y='Shucked weight',figsize=(9,9),color='black')
range=np.arange(0,2.5,0.1)
predictor=pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred= model.predict(X=predictor)
plt.plot(range,y_pred,color='b')
```
